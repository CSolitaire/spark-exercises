```python
import re

def password_check(df): 
    for i in range(len(df)):
        requirement = i[0]
        password = i[1]
        start = int(re.search(r'(^\d)', requirement).group())
        stop = int(re.search(r'-(\d.)', requirement).group(1))
        letter = str(re.search(r'([a-z])',requirement).group())
        return start, stop, letter, password

```
